Download the file growth_experiments.zipView in a new window, which contains phytoplankton fluorescence data (a proxy for biomass) from two growth experiments. In each experiment, Coscinodiscus diatoms, isolated from a station in the Amazon river plume, were grown in nutrient replete media.

- Plot the data from each experiment.

- Plot the log-transformed data from each experiment.

- Fit an exponential curve to the data to obtain the growth rate. Use the curve_fit function in scipy.optimize, which gives parameter estimates (popt) and their covariance (pcov). See help(curve_fit) for more details.

- Plot the fitted curve with the data

- Use the variance of the parameter estimates to calculate 95% confidence intervals for the growth rate in each experiment. Note that the standard error of the parameter popt[0] is given by the square root of pcov[0,0]. Assume a Student's t-distribution.

[This website was helpful in plotting the confidence intervals](http://kitchingroup.cheme.cmu.edu/blog/2013/02/12/Nonlinear-curve-fitting-with-parameter-confidence-intervals/)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
from scipy.optimize import curve_fit
from scipy.stats.distributions import t
```

```python
b7_1_filename = 'growth_experiments/B7-1.txt'
b7_2_filename = 'growth_experiments/B7-2.txt'

# Load the first experiment
df_1 = pd.read_csv(b7_1_filename, delimiter='\t',skiprows=[0])
df_1['Experiment'] = 'B7-1'

# Load the second experient
df_2 = pd.read_csv(b7_2_filename, delimiter='\t',skiprows=[0])
df_2['Experiment'] = 'B7-2'

# Concatinate dataframes
frames =  [df_1, df_2]
data = pd.concat(frames)
data.index = pd.to_datetime(data.Date)
data = data.drop('Date',1) #remove the Data str column
```

```python
exp1 = data.where(data.Experiment == "B7-1")
exp2 = data.where(data.Experiment == "B7-2")

# Exponential growth function
# where 
def func(x,a,b):
    return a * np.e ** (b * x)

# For B7-1
days1 = exp1['T'].dropna().values
RFU1 = exp1['RFU'].dropna().values
x_fit1 = np.linspace(days1.min()-1, days1.max()+2, 1000)
popt1, pcov1 = curve_fit(func, days1, RFU1)
y_fit1 = func(x_fit1, *popt1)

# For B7-2
days2 = exp2['T'].dropna().values
RFU2 = exp2['RFU'].dropna().values
x_fit2 = np.linspace(days2.min()-1, days2.max()+1, 1000)
popt2, pcov2 = curve_fit(func, days2, RFU2)
y_fit2 = func(x_fit2, *popt2)

# Confidence Intervals
#Interval = b +  SE * T
#DOF = 

alpha = 0.05
# B7-1
dof1 = len(RFU1)-len(popt1)
tval1 = t.ppf(1-(alpha/2), dof1)
sigma1 = pcov1[0,0]**.5
# B7-2
dof2 = len(RFU2)-len(popt2)
tval2 = t.ppf(1-(alpha/2), dof2)
sigma2 = pcov2[0,0]**.5

# Print Results
print "*" * 35
print "B7-1"
print "Growth rate: {0:.4f} ({1:.4f} - {2:.4f})".format(popt1[0],popt1[0] - (sigma1 * tval1),popt1[0] + (sigma1 * tval1))
print "Standard Error: {0:.4f}".format(sigma1)
print "T-test 95% confidence: {0:.4f}".format(tval1)


print "*" * 35
print "B7-2"
print "Growth rate: {0:.5f} ({1:.5f} - {2:.5f})".format(popt2[0],popt2[0] - (sigma2 * tval2),popt2[0] + (sigma2 * tval2))
print "Standard Error: {0:.4f}".format(sigma2)
print "T-test 95% confidence: {0:.4f}".format(tval2)
print "*" * 35
```

```python
fig, (ax0,ax1) = plt.subplots(2,1)
fig.set_size_inches(6,10)

ax0.plot(exp1['T'],exp1.RFU,'o',label='B7-1',markeredgecolor=None,ms=8)
ax0.plot(exp2['T'],exp2.RFU,'o',label='B7-2',markeredgecolor=None,ms=8)
ax0.plot(x_fit1,y_fit1,'b')
ax0.plot(x_fit2,y_fit2,'g')
ax0.legend(loc=2,numpoints=1)
ax0.set_xlabel('Days')
ax0.set_ylabel('RFU')

#Log transform
ax1.plot(exp1['T'],exp1.RFU,'o',label='B7-1',markeredgecolor=None,ms=8)
ax1.plot(exp2['T'],exp2.RFU,'o',label='B7-2',markeredgecolor=None,ms=8)
ax1.plot(x_fit1,y_fit1,'b')
ax1.plot(x_fit2,y_fit2,'g')
ax1.set_yscale('log')

ax1.legend(loc=2,numpoints=1)
ax1.set_xlabel('Days')
ax1.set_ylabel('RFU')

# Make plot the 95% confidence intervals
lower = [] # [a,mu (growth rate)]
upper = []
for p,var in zip(popt1, np.diag(pcov1)):
    sigma = var**0.5 
    lower.append(p - sigma*tval1)
    upper.append(p + sigma*tval1)

yfit1_low = func(x_fit1, *lower)
ax1.plot(x_fit1,yfit1_low,'b--')
yfit1_upper = func(x_fit1, *upper)
ax1.plot(x_fit1,yfit1_upper,'b--', label='95%')

lower = []
upper = []
for p,var in zip(popt2, np.diag(pcov2)):
    sigma = var**0.5 #Std-error
    lower.append(p - sigma*tval2)
    upper.append(p + sigma*tval2)
yfit2_low = func(x_fit2, *lower) 
ax1.plot(x_fit2,yfit2_low,'g--')
yfit2_upper = func(x_fit2, *upper)
ax1.plot(x_fit2,yfit2_upper,'g--', label='95%')

ax1.legend(loc=2,numpoints=1)
```
